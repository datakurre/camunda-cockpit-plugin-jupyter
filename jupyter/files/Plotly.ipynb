```python
import micropip
await micropip.install(['pandas', 'plotly==5.1.0'])

import plotly.graph_objects as go
import plotly.express as px
import pandas as pd
```

```python
import os
import js 
import json

BASE_URL = os.environ["CAMUNDA_ENGINE_API"]
```

```python
data = json.loads(await (await js.fetch(f"{BASE_URL}/history/process-instance")).text())
df = pd.DataFrame(data, columns=["id", "processDefinitionName", "startTime", "endTime", "durationInMillis"])
df = df[df.endTime.notna()]
df
```

```python
fig = px.scatter(df, x="startTime", y="durationInMillis", color="processDefinitionName",
                  title="Starts", labels={"processDefinitionName": "Process"})
fig.show()
```

```python
fig = px.timeline(df, x_start="startTime", x_end="endTime", y="id", color="processDefinitionName",
                  title="Timeline", labels={"processDefinitionName": "Process"})
fig.show()
```
